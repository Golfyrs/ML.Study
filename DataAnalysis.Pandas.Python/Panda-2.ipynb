```python
dataFrame = pandas.read_csv("avocado-prices.zip")

dataFrame = dataFrame.copy()[dataFrame['type'] == 'organic']
dataFrame['Date'] = pandas.to_datetime(dataFrame["Date"])

dataFrame.sort_values(by='Date', ascending=True, inplace=True)

dataFrame.head()
```

```python
graph_df = pandas.DataFrame()

regions = dataFrame['region'].unique()

for region in regions:
    region_df = dataFrame.copy()[dataFrame['region'] == region]
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}_price25ma'] = region_df['AveragePrice'].rolling(25).mean()

    if graph_df.empty:
        graph_df = region_df[[f'{region}_price25ma']]
    else:
        graph_df = graph_df.join(region_df[f'{region}_price25ma'])

graph_df.tail()
```

```python
graph_df.dropna().plot(figsize=(8,5), legend=False)
```
